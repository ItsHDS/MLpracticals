#**Matplotlib**

```python
population.head()
```

Histogram

```python
populationdata = population['Population (2020)']
countrydata = population['Country/Other']
areadata = population['Land Area (Km²)']
```

```python
plt.figure(figsize=(25,5))
plt.bar(countrydata, populationdata, color ='maroon',
        width = 0.4)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(25,5))
plt.bar(countrydata, areadata, color ='maroon',
        width = 0.4)
plt.xticks(rotation = 90)
plt.show()
```

```python
bins = [30, 40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(population['Country/Other'], bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```
